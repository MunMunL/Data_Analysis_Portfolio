#### Creating dummies and dropping first dummy variable

#### Grouping reasons for absence 

```python
data.columns.values
```

```python
reason_columns.columns.values
```

```python
data = data.drop(['Reason for Absence'], axis=1)
data
```

```python
reason_columns.loc[:,1:14].max(axis=1)
```

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
reason_type_1
```

#### Concatenate Column Values

```python
data = pd.concat([data, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
data
```

```python
data.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2' , 'Reason_3', 'Reason_4']
```

```python
data.columns = column_names
```

```python
data.head()
```

#### Reorder Columns

```python
column_names_reordered = ['Reason_1', 'Reason_2' , 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours' ]
data = data[column_names_reordered]
data.head()
```

#### Create a Checkpoint - reason modification

```python
data_reason_mod = data.copy()
```

```python
data_reason_mod
```

#### Date - extract month

```python
type(data_reason_mod['Date'][0])
```

```python
data_reason_mod['Date'] = pd.to_datetime(data_reason_mod['Date'],format = '%d/%m/%Y')
```

```python
type(data_reason_mod['Date'][0])
```

```python
data_reason_mod['Date'][0]
```

```python
data_reason_mod['Date'][0].month
```

```python
list_months = []
list_months
```

```python
data_reason_mod.shape
```

```python
for i in range(data_reason_mod.shape[0]):
    list_months.append(data_reason_mod['Date'][i].month)
    
```

```python
list_months
```

```python
len(list_months)
```

```python
data_reason_mod['Month Value'] = list_months
```

```python
data_reason_mod.head(20)
```

#### Date - extract day

```python
data_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
data_reason_mod['Day of the Week'] = data_reason_mod['Date'].apply(date_to_weekday)
data_reason_mod.head()
```

```python
data_reason_mod = data_reason_mod.drop('Date', axis=1)
```

```python
data_reason_mod 
```

```python
data_reason_mod.columns.values
```

```python
reordered_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
       'Month Value', 'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data_reason_mod = data_reason_mod[reordered_columns]
data_reason_mod
```

#### Create a Checkpoint - reason and date modification

```python
data_reason_date_mod = data_reason_mod.copy()
```
