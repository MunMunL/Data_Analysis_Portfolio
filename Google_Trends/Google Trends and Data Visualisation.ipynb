### Convert Strings to DateTime Objects

```python
print(f"Tesla's Month column data type is {type(df_tesla['MONTH'][0])}")
print(f"Unemployment Data's Month column data type is {type(df_unemployment['MONTH'][0])}")
print(f"Bitcoin's Search Month column data type is {type(df_btc_search['MONTH'][0])}")
print(f"Bitcoin's Price Date column data type is {type(df_btc_price['DATE'][0])}")
```

### Converting from Daily to Monthly Data

[Pandas .resample() documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html) <br>

```python
df_btc_monthly = df_btc_price.resample('M', on ='DATE').last()
```

```python
print(df_btc_monthly.shape)
df_btc_monthly.head()
```

# Data Visualisation

### Notebook Formatting & Style Helpers

```python
# Create locators for ticks on the time axis
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')
```

### Tesla Stock Price v.s. Search Volume

```python
plt.figure(figsize=(14,8), dpi = 200)
plt.title('Tesla Web Search vs Price', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

# Set separate axes
ax1=plt.gca()
ax2=ax1.twinx()

ax1.set_ylabel('TSLA Stock Price', color='g', size=14)
ax2.set_ylabel('Search Trend', color='red',size=14)

# Format ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

# Set axes limits
ax1.set_ylim([0,600])
ax2.set_ylim([0,40])
ax1.set_xlim([df_tesla['MONTH'].min(), df_tesla['MONTH'].max()])


ax1.plot(df_tesla['MONTH'], df_tesla['TSLA_USD_CLOSE'], color='g', linewidth=2)
ax2.plot(df_tesla['MONTH'], df_tesla['TSLA_WEB_SEARCH'], color='red', linewidth=2)

plt.show()
```

Tesla's increase in search trend follows the increase in stock prices with the spike in Mar 2016 coincides with when the Tesla Model 3 was released.

### Bitcoin (BTC) Price v.s. Search Volume

```python
plt.figure(figsize=(14,8), dpi = 200)
plt.title('Bitcoin News Search vs Resampled Price', fontsize=18)
plt.xticks(fontsize=14, rotation=45)

# Set separate axes
ax1=plt.gca()
ax2=ax1.twinx()

ax1.set_ylabel('BTC Price', color='g', size=14)
ax2.set_ylabel('Search Trend', color='red',size=14)

# Format ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

# Set axes limits
ax1.set_ylim([0,16000])
ax2.set_ylim([0,110])
ax1.set_xlim([df_btc_search['MONTH'].min(), df_btc_search['MONTH'].max()])

ax1.plot(df_btc_monthly.index, df_btc_monthly['CLOSE'], color='g', linewidth=2, linestyle='--')
ax2.plot(df_btc_search['MONTH'], df_btc_search['BTC_NEWS_SEARCH'], color='red', linewidth=2, marker='o')

plt.show()
```

Big increases in searches were closely accompanied by big increases in Bitcoin prices till it reached an all time high in 2018.  Bitcoin was a hot topic during 2017 and early 2017, so matched the high search volumes.
However during the next spike in price in 2019, the search trend didn't peak at the same rate as the price, likely due to Bitcoin being a fairly well known item/topic.

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

```python
plt.figure(figsize=(14,8), dpi = 200)
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate', fontsize=18)
plt.xticks(fontsize=14, rotation=45)
plt.grid(color='grey', linestyle='--', linewidth=1)

# Set separate axes
ax1=plt.gca()
ax2=ax1.twinx()

ax1.set_ylabel('FRED U/E Rate', color='g', size=14)
ax2.set_ylabel('Search Trend', color='red',size=14)

# Format ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

# Set axes limits
ax1.set_ylim([0,11])
ax2.set_ylim([0,110])
ax1.set_xlim([df_unemployment['MONTH'].min(), df_unemployment['MONTH'].max()])

ax1.plot(df_unemployment['MONTH'], df_unemployment['UNRATE'], color='g', linewidth=2, linestyle='--')
ax2.plot(df_unemployment['MONTH'], df_unemployment['UE_BENEFITS_WEB_SEARCH'], color='red', linewidth=2)

plt.show()
```

The spikes in searches for 'Unemployment benefit' generally happens at year end which shows potential seasonality in the job market.  
The financial crisis in 2008 caused a massive spike in unemployment.  
Between 2008 and 2014, the there were big spikes in searches not linked to increase in unemployment rates so there is likely another factor impacting the search trend.
The search volumes fluctuates month on month, a rolling average can help smooth it out to help identify other trends.


### 6-month rolling average 

```python
plt.figure(figsize=(14,8), dpi = 200)
plt.title('Rolling Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate', fontsize=18)
plt.xticks(fontsize=14, rotation=45)
plt.grid(color='grey', linestyle='--', linewidth=1)

# Set separate axes
ax1=plt.gca()
ax2=ax1.twinx()

ax1.set_ylabel('FRED U/E Rate', color='g', size=14)
ax2.set_ylabel('Search Trend', color='red',size=14)

# Format ticks
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

# Set axes limits
ax1.set_ylim([0,11])
ax2.set_ylim([0,90])
ax1.set_xlim([df_unemployment['MONTH'].min(), df_unemployment['MONTH'].max()])

# Calculate rolling 6 months
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()

ax1.plot(df_unemployment['MONTH'], roll_df['UNRATE'], color='g', linewidth=2, linestyle='--')
ax2.plot(df_unemployment['MONTH'], roll_df['UE_BENEFITS_WEB_SEARCH'], color='red', linewidth=2)

plt.show()
```

The chart for a 6 month rolling average indicates that the searches for 'Unemployment Benefit' happens before the unemployment rate goes up.  Similarly, the searched for 'Unemployment Benefit' goes down first before the unemplyment rate goes down.
